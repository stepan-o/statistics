##### Solving forwards, solving backwards

Probability is a _forwards problem_. It starts with a well-defined rule (like two-thirds of coin flips coming up heads) and then asks a question about the data that can be laid out in an exact way. While the result of an actual test may differ, we can exactly quantify what _should_ happen.

Statistics is a _backwards problem_. It starts from data and then asks **what was used to generate it**. For example, we might look at a series of 20 coin flips and try to determine if, given 15 heads and 5 tails, the coin might be fair, and if the coin is unfair, to what extent?

Starting from the end result and looking backwards can lead to many possible origin points. Nonetheless, with statistics we can mathematically quantify predictions, and to be a responsible statistician we also must quantify our uncertainty.

Even given these conditions, statistics can help find truths about data in astounding ways.

### Problem 1.1: judging fairness of a coin

Back to [section 1](#introduction) / Back to the [top](#top)

Maja has a coin which she thinks is unfair, but she isn't sure. She'd like to test the coin out by flipping it multiple times.

How many times does she need to flip the coin to be 100% certain of if the coin is fair?

#### Answer

Let $H_0$ be the null-hypothesis that the coin is fair.

Let $A_n$ be the event of flipping all heads on $n$ trials of a fair coin. Probability of flipping heads on each trial is $ \frac{1} {2} $, as it represents 1 out of 2 equally likely outcomes. 

As all trials represent independent events which all need to happen for $A_n$ to occur (**intersection**), so probability of that happening can be found using the rule of product:

$ P(A_n) = \left( \frac{1} {2} \right)^n  $

which would be greater than 0 for all positive $n$. 

Therefore, there is always some non-zero probability $ P(A_n) $ that $ H_0 $ is true, and we can never reject it will 100% certainty.

```python
max_trials = 1000

print("Probability of flipping all heads on 1'000 trials is: {0:.305f},\nor {0}".format((1 / 2) ** max_trials))

max_trials = 10

x = np.linspace(0, max_trials, 1000)
y = (1 / 2) ** x

f, ax = plt.subplots(1, figsize=(6, 6))
f.suptitle("There is always some non-null chance\nto flip any number of heads")
ax.plot(x, y)
ax.fill_between(x, 0, y, color='blue', alpha=0.2)
ax.text(1, 0.1, "$P(A_n)$")
ax.set_xlabel('Number of trials (n)')
ax.set_xticks(np.arange(max_trials + 1))
ax.set_ylabel('Probability of flipping all heads')
ax.axvline(0, color='black', linestyle='--', linewidth=1)
ax.axhline(0, color='black', linestyle='--', linewidth=1)
ax.set_xlim(-0.5, 10)
ax.grid(False)
plt.show()
```

#### Brilliant's explanation

No matter how fair or unfair a coin, as long as each side has a non-zero chance of showing up, there is always at least some probability for any given distribution.

For example, suppose the coin flips heads only once and tails 29 times. There are 30 ways this can happen (the heads can be any one of the 30 flips). In general, there are $2^{30} = 1073741824$ ways the coin sequence can happen, so the probability of 29 tails on a fair coin is $\large{\frac{30}{1073741824}}$, or about .0000028%. That's not the same as being 100% certain the coin isn't fair!

Despite this, we can often make a judgment call for truth, and simply declare some sort of cutoff. We might, for instance, want the condition that given the experiment was repeated in 100 different universes, in only one of them would our data be from random chance. (In mathematical terminology, this is the same as saying it has a $\large{p}$-value of .01, that is, 1%; more detail on $\large{p}$-values will come later in the course.)

### Problem 1.2: Judging fairness of a coin again

Let's step back and think on a smaller scale than the previous problem, but still in terms of a threshold where we feel some data is unlikely.

Maja thinks her coin is unfair. She flips it 4 times and gets heads every time. She calculates that this would only occur with a fair coin roughly 6% of the time. Can she conclude there is a roughly 94% chance that her coin is unfair?

#### Answer

Let $H_0$ be the null-hypothesis that the coin is fair.

Let $A_4$ be the event of a fair coin flipping heads 4 consequtive times. This would happen if results of four independent flips would all come out as heads (**intersection**). Therefore, probability of flipping 4 heads on a fair coin can be found using the rule of product:

$ P(A_4) = \left( \frac{1} {2} \right)^4 = \frac{1} {16} = 6.25\% $ 

The compliment of this event $ (A_4^c) $ is the event of a fair coin producing at least 1 tail on 4 trials and, according to the rule of compliment, it has a probability of 

$ P(A_4^c) = 1 - P(A_4) = 1 - \frac{1} {16}= \frac{15} {16} =  93.75\% $. 

If we would reject $ H_0 $ and call the coin unfair and the coin is fair, there would be a $ 6.25\% $ chance that we were wrong. But this is not the same as the probability of coin being unfair.

#### Brilliant's explanation

Correct answer: **No**

With the 6% chance mentioned in the problem, we are in one of those scenarios our fair coin gave unlikely data, so we reject the hypothesis it is fair and (wrongly) call it unfair. It is thus:

>The probability we would declare a fair coin to be unfair.

So the 94% part would be:

>The probability we avoided the scenario above (declaring a fair coin to be unfair).

This is the only thing we can conclude, and is different from stating outright there is a 94% chance the coin is unfair. More specifically, we also need to quantify the reverse scenario: our ability to avoid declaring an unfair coin to be fair. This is often not possible to do.

### Problem 1.3: judging fairness of a coin yet again

Back to [section 1](#introduction) / Back to the [top](#top)

For a more intuitive approach to the previous problem, we will reduce the situation to just two consecutive flips that come out as heads.

#### Answer

Let $ H_0 $ be the null-hypothesis that the coin is fair.

Let $ A_2$ be the event of flipping 2 consecutive heads on a fair coin. 

2 heads represent results of 2 independent trials, both of which need to come out as heads (**intersection**). For each trial of a fair coin, event of flipping heads represents 1 out of 2 equally likely outcomes, so probability of heads on each trial is $ \frac{1} {2} $. As both trials represent independent events, probability of heads on both trials can be found using the rule of product:

$ P(A_2) = \left( \frac{1} {2} \right)^2 = \frac{1} {4} = 25\% $

The compliment of this event $(A_2^c)$ is the event of a fair coin flipping at least 1 head on 2 trials. Its probability can be found using the rule of compliment:

$ P(A_2^c) = 1 - P(A_2) = 1 - \frac{1} {4} = \frac{3} {4} = 75\%$

If we would reject $ H_0 $, there would be a $ 25\% $ chance that a fair coin could have produced the result and we were wrong to reject the null-hypothesis. But this is not the same as the probability of the coin being unfair.

#### Brilliant's explanation

It seems absurd to claim this means there is a 75% chance the coin is fake! We can put a threshold on when we start to think data looks suspicious, but it is dependent on both judgment and context, and the threshold is not the same as the probability our conclusion is correct.

There are circumstances where data with a 5% chance of occurring is deemed significant, and there are times where the cutoff should be around 0.00006%. This is part of what makes backwards problems both so difficult and so intriguing.

### Problem 2.2: study of psychic abilities (unintentional deception)

Back to [section 2](#deception) / Back to the [top](#top)

A study is done on 250 people, where each person tries to predict whether a particular card from a shuffled deck is red or black before it is revealed. (The deck contains an equal number of each.)

Each participant makes 8 predictions. One participant gets all 8 predictions correct. The researcher claims that this person has exhibited psychic abilities.

Based on just the way the experiment is designed, is the researcher's claim a viable hypothesis? (Note we're not judging the plausibility of psychic ability in itself, just this particular experiment's design.)

#### Simulation

```python
%%time
max_guesses = 8
sample_size = 250
sims = 100000

# make 'max_guesses' guesses, return True if all guesses were correct
trial = lambda max_guesses: sum([np.random.randint(2) for i in np.arange(max_guesses)]) == max_guesses
# run 'sample_size' trials, return True if at least 1 of them was successful
sample_trial = lambda max_guesses, sample_size: True in [trial(max_guesses) for i in np.arange(sample_size)]
# run 'sims' sample trials
results = [sample_trial(max_guesses, sample_size) for i in np.arange(sims)]

# all samples where at least 1 trial guessed all cards right
successful_outcomes = sum(results)   
# total samples
total_outcomes = len(results)       
# probability of at least 1 person from a sample guessing all 8 cards right
probability = successful_outcomes / total_outcomes * 100
print("After {0:,} simulations, probability of at least 1 person out of the sample of {1:,} ".format(sims, sample_size) +
      "to guess all {0} cards right is {1:.2f}%".format(max_guesses, probability))
```

#### Solution

Let $ H_0 $ be the null-hypothesis that the person who guessed 8 cards right in a row has psychic abilities.

Let $ A_8 $ be the event of guessing 8 cards right in a row. As the deck contains an equal number of black and red cards, randomly guessing a card right each time represents 1 out of 2 equally likely outcomes, so probability of guessing a card right once is $ \frac {1} {2} $.

All of the guesses are assumed to be independent, and for 8 consecutive guesses, all of the individual guesses need to be right (**intersection**). Therefore, probability of 8 guesses being right can be found using the rule of product:

$ P(A_8) = \left( \frac{1} {8} \right) ^8 = \frac{1} {256} = 0.004\% $

The event $ (A_8^c) $ of not guessing 8 cards right (guessing at least 1 wrong) is the compliment of the event of guessing all 8 cards right. Therefore, its probability can be found using the rule of the compliment:

$ P(A_8^c) = 1 - P(A_8) = 1 - \frac{1} {256} = \frac{255} {256} = 99.6\% $

Let $ G_n $ be the event of at least 1 person out of the sample of $n$ to guess all 8 cards right. The compliment of this event  is the event $ (G_n^c) $ of none of the people making 8 correct guesses in a row. 

For none of the people to guess all cards right, each individual person needs to make at least 1 wrong guess. As was determined above, for one person probability of making at least 1 wrong guess is:

$ P(G_1^c) = \frac{255} {256} $

For a sample of $n$ people, all of them need to make at least 1 wrong guess (**intersection**). As guesses are assumed to be independent, probability of none of $n$ people get all cards right can be found using the rule of product:

$ P(G_n^c) = \left( \frac{255} {256} \right) ^n $

For a sample of 250 people, probability of none of them guessing all 8 cards correctly:

$ P(G_{250}^c) = \left( \frac{255} {256} \right) ^{250} = 37.6\% $

Therefore, probability of at least 1 person out of $n$ guessing all cards (based on random guesses) right, using the rule of compliment, is:

$ P(G_{250}) = 1 - P(G_{250}^c) = 1 - 37.6\% = 62.4 \% $

Since there is such a high probability of at least 1 person out of 250 guessing all cards right based on random guesses, **provided evidence does not offer sufficient support for the null-hypothesis** that the 1 person who made all right guesses has psychic abilities.

The larger the size of the sample, the higher is the probability that at least 1 person would guess all the cards right.

```python
(255 / 256) ** 250
```

```python
2 ** 8
```

```python
# limit of the graph
max_trials = 1000

# equation from analytical solution
x = np.linspace(0, max_trials, 1000)
y = (255 / 256) ** x

# plot the graph of the equation
f, ax = plt.subplots(1, figsize=(6, 6))
f.suptitle('With larger sample, at least 1 person will guess right')
ax.plot(x, y)
ax.fill_between(x, 0, y, color='red', alpha=0.2)
ax.fill_between(x, y, 1, color='blue', alpha=0.2)
ax.text(500, 0.65, "$P(G_n)$")
ax.text(300, 0.4, "At least 1 person from\nthe sample guessed\nall cards right")
ax.text(200, 0.1, "$P(G_n^c)$")
ax.set_xlabel('Size of the sample (n)')
ax.set_xticks(np.arange(0, max_trials + 1, max_trials / 10))
ax.set_ylabel('Probability of nobody guessing right')
ax.axvline(0, color='black', linestyle='--', linewidth=1)
ax.axhline(0, color='black', linestyle='--', linewidth=1)
ax.grid(False)
ax.set_xlim(-50, 1000)
plt.show()
```

#### Brilliant's explanation

Correct answer: **No**

If someone guesses randomly between two options, we would expect them to make 8 guesses correct with a probability of $\left(\frac{1}{2}\right)^{8} = \frac{1}{256}$. (Note this isn't exact if the cards are being removed from the deck, but it is close enough for approximation purposes.)

Given there are nearly 256 people in the experiment, it is reasonable to expect at least one of them would get all 8 correct by random chance. Therefore, the experiment as designed suggests nothing about psychic ability.

People attempting to fake numbers without being careful will have various tendencies crop up so that their values won't resemble random data.

This can be used to detect data fakery. If we just look at the last digits of a set of numbers (where rounding or some other bias isn't at play), the digits 0 through 9 should occur with roughly equal frequency. 

Remember to ask questions about context. Ask questions like: Do the statistics make sense for the situation? Is there something misleading in the presentation? Is there a bias in how the data is collected? Are categories combined in a deceptive way? Are we assuming anything that isn't true?

#### Defining function used to plot barcharts

```python
def plot_bars_with_minmax(series_to_plot, title, with_minmax=True, min_border=0.95, max_border=0.99, minmax_width=0.5):
    """
    function to plot a gray barchart from a pandas Series, 
    plots mean
    highlights bars with minimum values
    
    Input arguments: series_to_plot -- pandas Series -- Series to plot as a barchart
                     title          -- string        -- string containing title of the chart
                     with_minmax    -- boolean       -- option to highlight extreme values 
                                                        from the Series (default=True)
                     min_border     -- float         -- height limit for black min bar on the plot 
                                                        (default=0.95)
                     max_border     -- float         -- height limit for lightgray max bar on the plot 
                                                        (default=0.99)
                     minmax_width   -- float         -- width for min and max bars on the plot 
                                                        (default=0.5)
                     

    Output:          None, plots and shows barchart with mean and extremes (optional) highlighted
    """

    # create figure and axis
    f, ax = plt.subplots(1)

    # plot a gray bar chart from input Series
    plt.bar(x=series_to_plot.index,
            height=series_to_plot,
            color='gray')

    # plot mean of the series
    ax.axhline(series_to_plot.mean(), color='black', linestyle='--', linewidth=1)
    ax.text(0, 
            series_to_plot.mean() * 1.01, 
            "Mean of counts: {0:.2f}".format(series_to_plot.mean()))

    # optional: highlight the bars with minimum values (if input parameter 'with_min' is True)
    if with_minmax == True:
        min_se = series_to_plot[series_to_plot == series_to_plot.min()]
        plt.bar(x=min_se.index, 
                height=min_se * min_border, 
                color='black',
                width=minmax_width)
        
        max_se = series_to_plot[series_to_plot == series_to_plot.max()]
        plt.bar(x=max_se.index, 
                height=max_se * max_border, 
                color='lightgray',
                width=minmax_width)

    # set axis parameters
    ax.set_title(title)
    ax.set_xlabel("Number")
    ax.set_xticks(series_to_plot.index)
    ax.set_ylabel("Frequency of occurances")
    ax.tick_params('x', labelrotation=1, labelsize=20)
    ax.grid(False)
    plt.show()
```

Mean and median are both known as _measures of central tendency_.

It's worth noting that when people use the word "average" in everyday conversation, they _usually_ refer to the mean, but they also might intend one of the other measures of central tendency. Because of that ambiguity, we'll stick to the word "mean".

#### Mean

The most common way to measure the "center" of a set of data is with the _mean_, or average. It's the **sum of the values in the set divided by the number of items in the set.**

The mean of the data set below is 5 because $\frac{1+2+4+8+10}{5}=5$

* Mean: the mean of a set of numbers is given by the sum of the numbers in the set divided by the count of the numbers in the set.

<img src='img/stats/stat_mean_1.png' width=400>

```python
(1 + 2 + 4 + 8 + 10) / 5
```

```python
nums = [1, 2, 4, 8, 10]
stats.mean(nums)
```

##### Adding mean to a set does not change the value of its mean 

Let $\large{S}$ be the sum of terms of a dataset, $\large{n}$ be the number of terms, and $\large{M}$ be the mean. So $\large{\frac{S}{n}=M}$ which implies $\large{S=Mn}$.

Let's add another number, the original mean, to the set. We'll call $\large{Q}$ the new mean that results from doing this.

$\large{\frac{S + M}{n + 1} = Q}$

We want to justify $\large{Q=M}$ that is, the overall mean didn't change. Substituting $\large{S=Mn}$

$\large{\frac{Mn + M}{n + 1} = Q}$

and then factoring the $\large{M}$ out

$\large{\frac{M(n + 1)}{n + 1} = Q}$

lets us just cancel the $\large{(n + 1)}$ terms, leading us to $\large{M=Q}$.

So we can add the mean as many times as we want without changing the mean of the overall set.

#### Median

A second option for measuring the "center" of a set is the _median_, which is the value that divides a set in half when it is ordered from least to greatest; that is, the "middle number" in a set of values.

The median of the data set below is 8 because 8 is the middle number. Two values in the data set are smaller and two are larger.

* Median: the median of a set of numbers is given by the middle number of the set when it is written in numerical order; if there are an even number of items in the set, the median is the mean of the middle two numbers. (Note we must use a "sorted" set in order if we desire to find the median, but this is not necessary to find the mean.)

<img src='img/stats/stat_mean_2.png' width=400>

```python
nums = [5, 6, 8, 10, 15]
stat.median(nums)
```

##### Median of a set with an even number of terms

In a set with an even number of terms, the median is slightly more complicated, because there is no value in the set that is truly the "middle" term. When this happens, the median is calculated by finding the mean of the middle two terms.

What is the median of the set $A=\{2,2,3,4,5,6\}$?

```python
nums = [2, 2, 3, 4, 5, 6]
stat.mean(nums[2:4])
```

```python
stat.median(nums)
```

Correct answer: $3.5$

The set is already in order, so the median will be the mean of the middle two numbers: 3 and 4.

Thus, in this case, we have $median(A)=\frac{3+4}{2}=3.5$

#### Movement of mean and median

* Mean and median are close to each other in a balanced set.

* Median moves faster in a set that is unbalanced with no ouliers.

* Mean moves faster in a set that is unbalanced with outliers.

#### Defining function to plot sets, their mean, and median

```python
def plot_set_mean_median(set_to_plot, title, mean_lift=1.1, median_lift=1.1, legend_loc='best'):
    """
    function to plot a set, its mean, and median
    
    Input arguments: set_to_plot -- list   -- set of numbers to plot
                     title       -- string -- title of the plot
                     mean_lift   -- float  -- lift of text above mean line on the plot (default=1.1)
                     median_lift -- float  -- lift of text above median line on the plot (default=1.1)
                     legend_loc  -- string -- location of the legend for ax.legend (default='best')
    
    Output:          None, plots set, its mean, and median
    """
    # calculate mean and median of the input set
    set_mean = stats.mean(set_to_plot)
    set_median = stats.median(set_to_plot)
    
    # create a new axis
    ax = plt.axes(facecolor='whitesmoke')
    
    # plot the set
    ax.plot(set_to_plot, linestyle=':', linewidth=0.5, color='blueviolet', marker='o', label='Set numbers')
    
    # plot mean and median
    ax.axhline(set_mean, linestyle='--', linewidth=1.2, color='deeppink', label='Mean')
    ax.text(0, set_mean * mean_lift, "Mean of the set: {0:.2f}".format(set_mean))
    ax.axhline(set_median, linestyle=':', color='coral', label='Median')
    ax.text(0, set_median * median_lift, "Median of the set: {0:.2f}".format(set_median))
    
    # set axis parameters
    ax.set_title(title)
    ax.legend(loc=legend_loc, fancybox=True)
    plt.ylabel('Value')
    plt.xlabel('Element of the set')
    ax.grid(False)
    plt.show()
```

Then the actual price changes are -$\$$10, +$\$$30, +$\$$80, and +$\$$90, giving an average price change of +$\$$47.50, which is nearly 5% of the more expensive plans and nearly 50% of the smaller one. Obviously, grouping these plans together and averaging their changes as a percentage ignores meaningful difference between the plans.

In this case, Plan A is significantly cheaper than the other three plans; the price of the three has gone up by an average 6.67\%--more than double the reported 2.5\%. This way, adding Plan A to the other three results in the lower perceived average price.

### Calculating cab fare

#### Constructing inputs

* Three arrays with input values are provided: 
    * `'distance'` -- distance of the cab ride, in km 
    * `'wait_time'` -- wait time, in sec
    * `'four_pass'` -- whether there were four or more passengers in this trip (0-no, 1-yes)


* `'input_raw'` matrix is constructed from the three rows above, each row represents a variable


* Conversion factors are applied to raw input matrix `'input_raw'`, input matrix `'x'` is produced
    * `'distance'` is converted from m to km and divided by 143m (unit for price per distance)
    * `'wait_time'` is divided by 29 (unit for price per wait time)
    * `'four_pass'` is left unchanged


* input matrix `'x'` is transposed, each row represents a cab ride (instance)

```python
# input values
distance = [6, 6.7, 8.5, 3.8, 3.25, 4.5]
wait_time = [240, 225, 350, 180, 135, 275]
four_pass = [0, 0, 1, 0, 1, 0]

# conversion coefficients for inputs
distance_conversion = 1000 / 143 # distance is in km, fee is per 143m
wait_time_conversion = 1 / 29    # wait time is charged per 29 sec time intervals
four_pass_conversion = 1         # no conversion is applied to 'four_pass'

# input vector (raw)
# [0] distance travelled (in km), [1] wait time (in sec), 
# [2] four or more passengers(0 -- less than 4, 1 -- more than four)
input_raw = np.array([distance, wait_time, four_pass])

# conversion vector (distance * 1000 / 143, wait / 29, four_pass unchanged)
conversion = np.array([distance_conversion, wait_time_conversion, four_pass_conversion])

# input vector -- [0] trip distance in km (int), [1] wait time in seconds (int), [2] four or more passengers (binary)
# multiply each row (variable) of 'input_raw' by a corresponding converion
x = np.copy(input_raw)
for i in np.arange(len(input_raw)):
    x[i] = input_raw[i] * conversion[i]

# transposing the matrix for each row to represent an instance of a cab ride
x = x.T
print("Raw input (transposed):")
print(input_raw.T)
print("\nConverted input:")
print(x)
```

#### Applying a linear model to estimate cab fairs from inputs

* Array of coefficients $\beta$ is constructed (this array would be a product of a linear regression model)

* Output value `'y'` is produced as `'b_0'` (base fair) + dot product of input matrix `'x'` and vector of $\beta$ coefficients `'beta'`

```python
# vector of coefficients beta -- [0] fee per 143m of distance, fee per 29 sec of wait, fee for over 4 passangers) 
beta = np.array([0.25, 0.25, 2])
# base cab fair
beta_0 = 3.25

# y -- target -- cab fare
y = beta_0 + x.dot(beta)
y
```
